A partir de aqui comienza una linea en procesos internos a partir del inters de un prospecto de cliente:

$$
\textit{RFQ} \rightarrow \textit{Cotizaciones (CT)} \rightarrow \textit{Ordenes de compra (PO)} \rightarrow \textit{Factura (FC)}
$$

**RFQ**: Es el requerimiento que la persona interesada (precliente) necesita. Es la solicitud de servicio.
Esta RFQ tiene una fecha de vencimiento.

**CT (Cotizacin)**: Es un documento donde a partir de los requerimientos del precliente son solventados por la empresa, ofrecemos un costo total por el "proyecto" el cual resuelve los requerimientos que tiene el solicitante. Esta cotizacin incluyen los gastos totales del servicio y puede estar dividido en partidas o "subproyectos";

$\textit{RFQ}[id] \rightarrow \textit{CT}[id_{rfq}]$

Si la cotizacin es aceptada, entonces la CT tendr asociada una PO.

$\textit{CT}[id_{po}] \rightarrow \textit{PO}[id]$ .

**PO (Orden de compra)**: La orden de compra es el documento de aceptacin de proyecto; a partir de la cotizacin generada y enviada el cliente acepta el costo de cotizacin y procede a *confirmar la realizacin del servicio*
Toda PO requiere de tener una FC, en dado caso de que no tenga es importante denotarlo.

$\textit{PO}[id_{fc}] \rightarrow \textit{FC}[id]$.

**FC (Factura)** : El documento de facturacin representa el trmino del proyecto o subproyecto, en donde la empresa emite el monto facturado por el proyecto con monto, fecha y nombre a quien se emitir la factura
Las facturaciones tienen procesos donde la emisin de FC puede ser instantanea o con retardo.

Este proceso representa la columna vertebral en funcin del tiempo del servicio que ofrece la empresa.
